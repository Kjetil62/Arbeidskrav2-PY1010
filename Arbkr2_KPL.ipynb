b) Lag et program som ber brukeren skrive inn et land (eksempelvis England). Programmet skal på bakgrunn av dette skrive ut følgende setning: London er hovedstaden i England og det er 8.982 mill. innbyggere i London.


```python
data = {
    "Norge": ["Oslo", 0.634],
    "England": ["London", 8.982],
    "Frankrike": ["Paris", 2.161],
    "Italia": ["Roma", 2.873]
    }

def finn_hoveds_befolk(land): # Avsnittet finner hovedstad, befolkning i oppgitt land eller returnerer at programet ikke har info om landet
    if land in data:
        hovedstad, befolkning = data[land]
        print(f'{hovedstad} er hovedstaden i {land} og det er {befolkning} mill, innbyggere i {hovedstad}.')
    else:
        print(f'Beklager, jeg har ingen informasjon om {land}.')

land = input('Skriv inn et land: ')  # Input land
finn_hoveds_befolk(land)
```

4c) Lag et program som ber brukeren skrive inn info om et nytt land (altså et land som ikke allerede finnes i dictionaryen data). Videre skal brukeren oppgi hovedstad og antall innbyggere for det «nye» landet. Programmet skal så utvide/oppdatere dictionaryen med den nye informasjonen. Dictionaryen data skrives så til skjerm. 

```python
data = {
    'Norge': ['Oslo', 0.634],
    'England': ['London', 8.982],
    'Frankrike': ['Paris', 2.161],
    'Italia': ['Roma', 2.873]
    }

def finn_hoveds_befolk(land): # Avsnittet finner hovedstad, befolkning i oppgitt land eller returnerer at programet ikke har info om landet og gir muliget for å legge til ny informasjon
    if land in data:
        hovedstad, befolkning = data[land]
        print(f'{hovedstad} er hovedstaden i {land} og det er {befolkning} mill, innbyggere i {hovedstad}.')
    else:
        print(f'Beklager, jeg har ingen informasjon om {land}, legg til informasjon om land, hovedstad og befolkning i databasen.')
        legg_til_land(land)

def legg_til_land(land):  # input til info om nytt land
    hovedstad = input(f'Skriv inn hovedstaden i {land}: ')
    befolkning = float(input(f'Skriv inn befolkningen i {hovedstad} (i millioner): '))
    data[land] = [hovedstad, befolkning]
    print(f'{land} er lagt til i databasen med hovedstaden {hovedstad} og en befolkning på {befolkning} millioner.') 

    print(f'Ny dictonary:')  # Legger nytt land, hovedstad og befolkning i dictonary
    print(f'data = {data}')  # Sriver ut ny dictonary

land = input(f'Skriv inn et land: ')  # Input, nytt land
finn_hoveds_befolk(land)
```

Oppg 5)  Lag et program med en funksjon som tar a og b som inn-argumenter og som så regner ut arealet og «ytre» omkrets til en figur satt sammen av en rettvinklet trekant og en halvsirkel, se figuren under.  Med «ytre» omkrets menes samlet lengde av de sorte strekene. Funksjonen skal returnere arealet og «ytre» omkrets, som så skrives til skjerm med passende tekst. 

```python
import math 

def beregn_areal_og_omkrets(a, b,): # Beregn areal og omkrets
    
    triangel_omr = 0.5 * a * b  # avsnittet beregner areal
    radius = a / 2 
    halv_sirk_omr = 0.5 * math.pi * radius**2
    totalt_område = triangel_omr + halv_sirk_omr
   
    hyp = math.sqrt(a**2 + b**2)  # Avsnittet beregner omkrets
    triangel_omkr_u_a = b + hyp
    halvsirk_omkr = halv_sirk_omr
    total_omkr = triangel_omkr_u_a + halvsirk_omkr
    return totalt_område, total_omkr

a = float(input('Skriv inn lengden på a: '))  # Input lengden på a
b = float(input('Skriv inn lengden på b: '))  # Input lengden på b

totalt_område, total_omkr = beregn_areal_og_omkrets(a, b) 

print(f'Arealet av figuren er: {totalt_område:.4f}')
print(f'Ytre omkrets av figuren er: {total_omkr:.4f}')
```

Oppg 6) Skriv en kode som plotter funksjonen 𝑓(𝑥) = −𝑥2 − 5, for x på intervallet [-10,10]. 

```python
import numpy as np

import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 200)  # Generer 200 punkter av x
f_av_x = -x**2 - 5  # Generer 200 punkter av Y

plt.plot(x, f_av_x) # Plot funksjonen

plt.title('Plot av funksjonen f(x) = -x^2 - 5')  # avsnittet legger til tittel og etiketter
plt.xlabel('x verdi')
plt.ylabel('f(x) verdi')

plt.show()  # Vis plottet
```
